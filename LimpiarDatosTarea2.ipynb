```python
#copiar dataframe
df=df.copy()
```

```python
df_copy= df.copy()
```

```python
#quitamos con drop los datos con dimenciones=0
df_copy = df_copy.drop(df_copy[df_copy["x"]==0].index)
df_copy = df_copy.drop(df_copy[df_copy["y"]==0].index)
df_copy = df_copy.drop(df_copy[df_copy["z"]==0].index)
df_copy.shape
```

```python
#validamos las filas donde hay 0
df_copy[df_copy.apply(lambda row: (row == 0).any(), axis=1)]
```

```python
df_copy.apply(lambda col: (col == 0).sum())
#podemos notar que ya no hay columnas con valores 0
```

```python
#vamos a quitar los valores anormales que excedan nuestros maximos y minimos # Asumiendo que df_copy ya ha sido definido previamente
df_copy = df_copy[(df_copy["depth"] < 75) & (df_copy["depth"] > 45)]
df_copy = df_copy[(df_copy["table"] < 80) & (df_copy["table"] > 40)]
df_copy = df_copy[(df_copy["x"] < 30)]
df_copy = df_copy[(df_copy["y"] < 30)]
df_copy = df_copy[(df_copy["z"] < 30) & (df_copy["z"] > 2)]
df_copy.shape
```

```python
#Hacer una lista de las varibles categoricas
s = (df_copy.dtypes =="object")
object_cols = list(s[s].index)
print("Variable categoricas:")
print(object_cols)
```

```python
#Usamos LabelEncoder para hacer las categorias de las columnas cut, color y
#clarity para su clasificacion
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
for col in object_cols:
    df_copy[col] = label_encoder.fit_transform(df_copy[col])
df_copy.head()
```
